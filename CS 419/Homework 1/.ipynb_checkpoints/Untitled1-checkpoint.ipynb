```python
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('./train.csv')
MAX_DEPTH = 250
MIN_NODE_LENGTH = 1
data.head()
```

```python
def normalize(mat):
    a = np.array(mat)
    a = a.astype(np.float64)
    b = np.apply_along_axis(lambda x: (x - np.min(x)) /
                            float(np.max(x) - np.min(x)), 0, a)
    return b
```

```python
data = np.array(data).astype('float64')
np.random.shuffle(data)

split = int(0.69 * data.shape[0])

X = data[:split,:-1]
y = data[:split,-1]

data_val = data[split:,:]

leaves = []
y = y.reshape(-1,1)
X = normalize(X)
```

```python
X_test = np.array(pd.read_csv('test.csv'))
X_test = normalize(X_test)
model = DecisionTreeRegressor()
model.fit(X,y)
y_pred = model.predict(X_test)

f = open('output1.csv','w')
f.write('Id,quality\n')
for i in range(X_test.shape[0]):
    f.write(str(i+1)+','+str(y_pred[i])+'\n')
f.close()
```
